```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
K = np.array([16*33*33,12*33*33,8*33*33,4*33*33,2*33*33,1*33*33,11*33, 3*33])
error = np.array([0.206251256167888,0.37711838632822,0.500494999811053,0.501480426639318,0.443961611017584,
0.517864890396595,0.50387561507523, 0.5906662493944168])
```

```python
#polynomial fit with degree = 2
model = np.poly1d(np.polyfit(K, error, 3))

#add fitted polynomial line to scatterplot
polyline = np.linspace(0, 17500, 50)
```

```python
plt.scatter(K, error,s=60)
plt.plot(polyline, model(polyline),'r--')
plt.xlabel('k')
plt.ylabel('Frobenius Error')
plt.title('Average K-Latent SAE Error')
plt.savefig('../images/cnn_sae_error_plot.jpg', dpi=200)
plt.show()
```
